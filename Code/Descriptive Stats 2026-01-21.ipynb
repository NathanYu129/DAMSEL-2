```python
def histoplot(input, title: str, xlim = 14, ylim = 800):
    try:
        input = input.drop('Label', axis=1)
    except:
        pass
    ax = plt.gca()
    #ax.set_xlim((0,14))
    ax.set_xlim((0,xlim))
    #ax.set_ylim((0,800))
    ax.set_ylim((0,ylim))
    #input = input.iloc[1:]
    flattened = input.to_numpy().flatten()
    plt.title(f'Histogram of {title} values')
    n, bins, patches = plt.hist(flattened, bins= [i/2 for i in range(0,30,1)])
    print(n)
    plt.show()

def doublehistoplot(input1, input2, title: str, xlim = 14, ylim = 800):
    try:
        input1 = input1.drop('Label', axis=1)
    except:
        pass
    try:
        input2 = input2.drop('Label', axis=1)
    except:
        pass
    ax = plt.gca()
    #ax.set_xlim((0,14))
    ax.set_xlim((0,xlim))
    #ax.set_ylim((0,800))
    ax.set_ylim((0,ylim))
    #input = input.iloc[1:]
    flattened1 = input1.to_numpy().flatten()
    flattened2 = input2.to_numpy().flatten()
    #plt.title(f'Histogram of {title} values')
    plt.title(title)
    n, bins, patches = plt.hist(flattened1, bins= [i/2 for i in range(0,30,1)])
    n, bins, patches = plt.hist(flattened2, bins= [i/2 for i in range(0,30,1)])
    print(n)
    plt.show()
```
